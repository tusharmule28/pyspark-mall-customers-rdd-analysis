### Transformation- Wide and Narrow


```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Transformation- Wide and Narrow") \
    .getOrCreate()
```

```python
csv_path = "hdfs:///data/Mall_Customers.csv"
```

```python
rdd = spark.sparkContext.textFile(csv_path)
```

```python
header = rdd.first()
```

```python
data_rdd = rdd.filter(lambda row: row != header)
```

```python
data_rdd.take(5)
```

```python
mapped_rdd = data_rdd.map(lambda line: line.split(","))
```

```python
active_rdd = mapped_rdd.filter(lambda row: row[-1] == "True")
```

```python
active_rdd.take(5)
```

```python
state_pairs = mapped_rdd.map(lambda row: (row[3], 1))
```

```python
state_counts = state_pairs.reduceByKey(lambda a, b: a + b)
```

```python
results = {
    "Total Rows": data_rdd.count(),
    "First Row": data_rdd.first(),
    "Take 3 Rows": data_rdd.take(3),
    "Mapped Rows (first 5)": mapped_rdd.take(5)
}

results
```
